```python
merged_df = merged_df.dropna(subset=['Review Decision', 'Umpire Decision']).copy()

# Fill missing numerical values using `.loc[]`
numeric_cols = merged_df.select_dtypes(include=['number']).columns
merged_df.loc[:, numeric_cols] = merged_df[numeric_cols].fillna(merged_df[numeric_cols].mean())

# Fill missing categorical values using `.loc[]`
categorical_cols = merged_df.select_dtypes(include=['object']).columns
merged_df.loc[:, categorical_cols] = merged_df[categorical_cols].fillna('Unknown')
```

```python
import scipy.stats as stats

# Define target variable and categorical features
target_var = 'Umpire Decision'
categorical_features = merged_df.select_dtypes(include=['object', 'category']).columns.tolist()
categorical_features.remove(target_var)  # Remove target variable if in the list

# Perform Chi-Square test for each categorical variable
chi2_results = {}
for feature in categorical_features:
    contingency_table = pd.crosstab(merged_df[feature], merged_df[target_var])
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
    chi2_results[feature] = {'Chi2 Statistic': chi2, 'p-value': p, 'Degrees of Freedom': dof}

# Convert results to DataFrame and display
chi2_df = pd.DataFrame.from_dict(chi2_results, orient='index')
print(chi2_df.sort_values(by='p-value'))
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------
# Handling Missing Values

df = merged_df

# -------------------------
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values[missing_values > 0])

# Replacing the missing values: Median for numerical, Mode for categorical
def impute_missing_values(df):
    for column in df.columns:
        if df[column].dtype == 'object':
            df[column].fillna(df[column].mode()[0], inplace=True)  # Replace categorical missing values with mode
        else:
            df[column].fillna(df[column].median(), inplace=True)  # Replace numerical missing values with median
    return df

df = impute_missing_values(df)

# -------------------------
# Categorical Feature Analysis (Chi-Square Test)
# -------------------------
target_var = 'Umpire Decision'
categorical_features = df.select_dtypes(include=['object', 'category']).columns.tolist()
categorical_features.remove(target_var)

chi2_results = {}
for feature in categorical_features:
    try:
        contingency_table = pd.crosstab(df[feature], df[target_var])
        chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
        chi2_results[feature] = {'Chi2 Statistic': chi2, 'p-value': p, 'Degrees of Freedom': dof}
    except ValueError as e:
        print(f"Skipping {feature}: {e}")

chi2_df = pd.DataFrame.from_dict(chi2_results, orient='index')
print("Chi-Square Test Results:")
print(chi2_df.sort_values(by='p-value'))

# -------------------------
# Visualization: Chi-Square Test Results
# -------------------------
plt.figure(figsize=(12, 6))
sns.barplot(y=chi2_df.index, x=chi2_df['p-value'], palette='coolwarm')
plt.axvline(x=0.05, color='r', linestyle='--', label='Significance Threshold (0.05)')
plt.xlabel('p-value')
plt.ylabel('Categorical Variables')
plt.title('Chi-Square Test Results')
plt.legend()
plt.show()
```

```python
# Convert 'Umpire Decision' to numerical representation
decision_mapping = {'right': 1, 'wrong': 0}
df['Umpire_Decision_Numeric'] = df['Umpire Decision'].map(decision_mapping)

# Select numerical features including the new numeric 'Umpire Decision'
numerical_features = df.select_dtypes(include=np.number).columns.tolist()

# Calculate the correlation matrix
correlation_matrix = df[numerical_features].corr()

# Create the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features including Umpire Decision')
plt.show()
```
